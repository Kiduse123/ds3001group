```python
# Packages

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read in files

outcomes = pd.read_excel("WinningPartyByCounties.xlsx") # party outcome at county level

rawdemo2012 = pd.read_excel("demographicsallstates2012.xlsx") # demographic variables at county level 2012
rawdemo2016 = pd.read_excel("demographicsallstates2016.xlsx") # demographic variables at county level 2016
rawdemo2020 = pd.read_excel("demographicsallstates2020.xlsx") # demographic variables at county level 2020

papoverty = pd.read_csv("papoverty.csv") # poverty variable at county level -- PA
papoverty = papoverty.drop([0,1, 38, 39]) # Drop national and state level statistics
```

```python
# Filter files for Pennsylvania

paoutcomes = outcomes[outcomes['state'] == 'PENNSYLVANIA'] # filtered outcome variable for PA

demo2012 = rawdemo2012[rawdemo2012['State Name'] == 'Pennsylvania']
demo2016 = rawdemo2016[rawdemo2016['State Name'] == 'Pennsylvania']
demo2020 = rawdemo2020[rawdemo2020['State Name'] == 'Pennsylvania'] # filtered demographic variables for PA
```

```python
# Separate outcomes & poverty by year

paoutcomes2012 = paoutcomes[paoutcomes['year'] == 2012]
paoutcomes2016 = paoutcomes[paoutcomes['year'] == 2016]
paoutcomes2020 = paoutcomes[paoutcomes['year'] == 2020] # separated outcomes variable by year

papoverty2012 = papoverty[papoverty['Year'] == 2012]
papoverty2016 = papoverty[papoverty['Year'] == 2016]
papoverty2020 = papoverty[papoverty['Year'] == 2020] # separated poverty variable by year
```

```python
# Merge

election2012 = pd.merge(demo2012, papoverty2012, left_on='County Name', right_on='Name', how='outer') # Merge poverty with demographics
election2012 = pd.merge(election2012, paoutcomes2012, left_on='ID', right_on='county_fips', how='outer') # Merge outcomes with election2012
# now have a complete file for 2012

election2016 = pd.merge(demo2016, papoverty2016, left_on='County Name', right_on='Name', how='outer') # Merge poverty with demographics
election2016 = pd.merge(election2016, paoutcomes2016, left_on='ID', right_on='county_fips', how='outer') # Merge outcomes with election2016
# now have a complete file for 2016

election2020 = pd.merge(demo2020, papoverty2020, left_on='County Name', right_on='Name', how='outer') # Merge poverty with demographics
election2020 = pd.merge(election2020, paoutcomes2020, left_on='ID', right_on='county_fips', how='outer') # Merge outcomes with election2012
# now have a complete file for 2020
```
